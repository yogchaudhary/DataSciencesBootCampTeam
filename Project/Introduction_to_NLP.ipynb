```python
# TWO PARTS OF EMBEDDINGS OR NUMERICAL REPRESENTATION
#1) TFIDF,COUNT VECTORIZERS such static models dont need tokenization
#2) Models like bert,Doc2Vec need # Tokenization and removing unwanted characters
```

```python
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
def clean_text(sentence):

  # Converting all words to lower
  sentence_lowered=sentence.lower()
  # fetched words from nltk package
  stop = stopwords.words('english')

  stop_words_removed_sentence=[]
  for k in sentence_lowered.split(" "):
    if k not in stop:
      stop_words_removed_sentence.append(lemmatizer.lemmatize(k))
  stop_words_removed_sentence=" ".join(stop_words_removed_sentence)
  print(stop_words_removed_sentence)
```

```python
text_sample="I am so angry that i made this post available"
clean_text(text_sample)
```

```python
from sentence_transformers import SentenceTransformer
# BERT covered for 100 languages different models you can read about them

model = SentenceTransformer('all-MiniLM-L6-v2')
sentences = ['This framework generates embeddings for each input sentence',
    'Sentences are passed as a list of string.',
    'The quick brown fox jumps over the lazy dog.']

#Sentences are encoded by calling model.encode()
embeddings = model.encode(sentences)

#Print the embeddings
for sentence, embedding in zip(sentences, embeddings):
    print("Sentence:", sentence)
    print("Embedding:", embedding.shape)
    print("")
```
