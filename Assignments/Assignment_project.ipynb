```python
# 1
# Count the number of properties in each neighborhood group
property_count = df['neighbourhood_group'].value_counts()

# Display the count of properties in each neighborhood group
print(property_count)
```

```python
# 2
# Group the data by neighborhood group and calculate the mean price for each group
grouped_data = df.groupby('neighbourhood_group')['price'].mean()

# Find the neighborhood group with the highest mean price
highest_price_group = grouped_data.idxmax()

# Get the highest mean price
highest_price = grouped_data.max()

# Display the result
print("Neighborhood group with the highest prices:", highest_price_group)
print("Highest mean price:", highest_price)
```

```python
# 3
# Group the data by neighborhood group and calculate the mean price for each group
grouped_data = df.groupby('neighbourhood_group')['price'].mean()

# Find the neighborhood group with the lowest mean price
lowest_price_group = grouped_data.idxmin()

# Get the lowest mean price
lowest_price = grouped_data.min()

# Display the result
print("Neighborhood group with the lowest prices:", lowest_price_group)
print("Lowest mean price:", lowest_price)
```

```python
# 4
# Group the data by neighborhood group and find the highest and lowest prices for each group
grouped_data = df.groupby('neighbourhood_group')['price'].agg(['max', 'min'])

# Display the result
print("Highest and Lowest Prices for Airbnb in Each Neighborhood Group:")
print(grouped_data)
```

```python
#  4
# Group the data by Neighborhood Group and find the highest and lowest prices
grouped_data = df.groupby('neighbourhood_group')
highest_prices = grouped_data['price'].max()
lowest_prices = grouped_data['price'].min()

# Plotting the data
fig, ax = plt.subplots(figsize=(10, 6))

# Plotting the highest prices
ax.bar(highest_prices.index, highest_prices.values, color='blue', alpha=0.7, label='Highest Price')

# Plotting the lowest prices
ax.bar(lowest_prices.index, lowest_prices.values, color='green', alpha=0.7, label='Lowest Price')

# Adding labels and title
ax.set_xlabel('Neighborhood Group')
ax.set_ylabel('Price')
ax.set_title('Highest and Lowest Prices by Neighborhood Group')
ax.legend()

# Rotating the x-axis labels if needed
plt.xticks(rotation=45)

# Display the bar chart
plt.show()
```

```python
# 4  Highest and lowest price airbnb in each Neighbour Hood Group

# Group the data by Neighborhood Group and find the highest and lowest prices
grouped_data = df.groupby('neighbourhood_group')
highest_prices = grouped_data['price'].max()
lowest_prices = grouped_data['price'].min()

# Iterate over each Neighborhood Group and print the highest and lowest prices
for group, highest_price in highest_prices.items():
    lowest_price = lowest_prices[group]
    print("Neighborhood Group:", group)
    print("Highest Price:", highest_price)
    print("Lowest Price:", lowest_price)
    print()

```

```python
# 5 Which Host has Highest count of properties

# Count the number of properties for each host
host_counts = df['host_id'].value_counts()

# Find the host with the highest count of properties
highest_count_host = df[df['host_id'] == host_counts.index[0]]['host_name'].iloc[0]

# Print the result
print("Host with the highest count of properties:", highest_count_host)

```

```python
# 6 Which neighbourhood is most expensive and cheapest

# Find the most expensive and cheapest neighborhoods
most_expensive_neighborhood = df.loc[df['price'].idxmax(), 'neighbourhood']
cheapest_neighborhood = df.loc[df['price'].idxmin(), 'neighbourhood']

# Print the results
print("Most Expensive Neighborhood:", most_expensive_neighborhood)
print("Cheapest Neighborhood:", cheapest_neighborhood)


```

```python
# 7

# Count the occurrences of each room_type
room_type_counts = df['room_type'].value_counts()

# Create a pie chart
fig, ax = plt.subplots()

# Plotting the pie chart
ax.pie(room_type_counts, labels=room_type_counts.index, autopct='%1.1f%%', startangle=90)

# Adding title
ax.set_title('Room Type Distribution')

# Display the pie chart
plt.show()
```

```python
# 8

# Group the data by neighborhood group
grouped_data = df.groupby('neighbourhood_group')

# Create a scatter plot for each neighborhood group
fig, ax = plt.subplots(figsize=(10, 6))

# Define colors and markers for each neighborhood group
colors = ['red', 'blue', 'green', 'orange', 'purple']
markers = ['o', 's', '^', 'D', 'v']

# Iterate over each neighborhood group and plot scatter points
for i, (group, data) in enumerate(grouped_data):
    ax.scatter(data['longitude'], data['latitude'], color=colors[i], marker=markers[i], label=group)

# Add labels and title
ax.set_xlabel('Longitude')
ax.set_ylabel('Latitude')
ax.set_title('Scatter Plot of Latitude vs Longitude for Different Neighborhood Groups')
ax.legend()

# Display the scatter plot
plt.show()
```

```python
# 9

# Create a scatter plot for each room type
fig, ax = plt.subplots(figsize=(10, 6))

# Define colors and markers for each room type
colors = ['red', 'blue', 'green', 'orange']
markers = ['o', 's', '^', 'D']

# Iterate over each room type and plot scatter points
for i, room_type in enumerate(df['room_type'].unique()):
    room_type_data = df[df['room_type'] == room_type]
    ax.scatter(room_type_data['longitude'], room_type_data['latitude'], color=colors[i], marker=markers[i], label=room_type)

# Add labels and title
ax.set_xlabel('Longitude')
ax.set_ylabel('Latitude')
ax.set_title('Scatter Plot of Latitude vs Longitude for Different Room Types')
ax.legend()

# Display the scatter plot
plt.show()

```

```python
# 11


# Calculate mean, highest, and lowest prices for each room type
room_type_stats = df.groupby('room_type')['price'].agg(['mean', 'max', 'min'])

# Set up the bar chart
fig, ax = plt.subplots(figsize=(10, 6))

# Set the positions and width for the bars
bar_positions = range(len(room_type_stats.index))
bar_width = 0.2

# Plotting the bars for mean prices
ax.bar(bar_positions, room_type_stats['mean'], width=bar_width, color='blue', alpha=0.7, label='Mean Price')

# Plotting the bars for highest prices
ax.bar([p + bar_width for p in bar_positions], room_type_stats['max'], width=bar_width, color='green', alpha=0.7, label='Highest Price')

# Plotting the bars for lowest prices
ax.bar([p + (2 * bar_width) for p in bar_positions], room_type_stats['min'], width=bar_width, color='red', alpha=0.7, label='Lowest Price')

# Adding labels and title
ax.set_xlabel('Room Type')
ax.set_ylabel('Price')
ax.set_title('Comparison of Room Types with Mean, Highest, and Lowest Prices')
ax.set_xticks([p + bar_width for p in bar_positions])
ax.set_xticklabels(room_type_stats.index)
ax.legend()

# Display the bar chart
plt.show()
```
