```python
# OPENCV READ IMAGES

# Python code to read image
import cv2
from google.colab.patches import cv2_imshow

# OPENCV ALWAYS USES NUMPY ARRAY TO READ IMAGES
img = cv2.imread("appleimage.jpg")
# height and width
print(img.shape)
cv2_imshow(img)
#cv2.imshow(nameofimage,image)
```

Syntax: cv2.imread(path, flag)

Parameters:
path: A string representing the path of the image to be read.
flag: It specifies the way in which image should be read. Its default value is cv2.IMREAD_COLOR

Return Value: This method returns an image that is loaded from the specified file.

Note:
The image should be in the working directory or a full path of image should be given.
By default, OpenCV stores colored images in BGR(Blue Green and Red) format.

```python
#import cv2, numpy and matplotlib libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt

img=cv2.imread("appleimage.jpg")
# Converting BGR color to RGB color format

BGR_img = cv2.cvtColor(img, cv2.COLOR_RGB2BGR)
print(BGR_img.shape)
#Displaying image using plt.imshow() method
cv2_imshow(BGR_img)
```

```python

RGB_img = cv2.cvtColor(BGR_img,cv2.COLOR_BGR2RGB)
cv2_imshow(RGB_img)
```

```python
import cv2

img = cv2.imread("appleimage.jpg", cv2.IMREAD_GRAYSCALE)
print(img.shape)
cv2_imshow(img)
```

```python
import cv2

img = cv2.imread("appleimage.jpg")
img_bigger=cv2.resize(img, (780, 540))
print(img_bigger.shape)
cv2_imshow(img_bigger)
```

```python
import cv2

img = cv2.imread("appleimage.jpg")
start_point=(1,1)
end_point=(256,192)

# FORMAT IS BGR
color=(255,0,0)
thickness=2

image=cv2.rectangle(img,start_point,end_point,color,thickness)
cv2_imshow(image)
```

```python
import cv2

img = cv2.imread("appleimage.jpg")
image_flipped=cv2.flip(img,-1)
cv2_imshow(image_flipped)
```

```python
import cv2
img = cv2.imread("appleimage.jpg")
#cv2.ROTATE_90_COUNTERCLOCKWISE
#cv2.ROTATE_180
image_ROTATE=cv2.rotate(img,cv2.ROTATE_90_CLOCKWISE)
cv2_imshow(image_ROTATE)
```

```python
import cv2
img = cv2.imread("appleimage.jpg")
kernel_size=(4,2)
image_blur=cv2.blur(img,kernel_size)
cv2_imshow(image_blur)
```
